## Data Cleaning

Checking missing values

```python
df.isnull().sum()
```

Handling missing values

```python
df=df.dropna()
df
```

Rechecking missing values

```python
df.isnull().sum()
```

Checking duplicates

```python
df.duplicated().sum()
```

Removing duplicates

```python
df=df.drop_duplicates()
df
```

Rechecking duplicates

```python
df.duplicated().sum()
```

Convert 'Date' column to datetime format

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

## Data Analysis Tasks

What is the average Close price of stocks

```python
df['Close'].mean()
```

Average Close price per Sector

```python
df.groupby('Sector')['Close'].mean().plot(kind='bar')
```

Which Symbol has the highest Market_Cap

```python
df.groupby('Symbol')['Market_Cap'].sum().plot(kind='barh')
```

```python
df.loc[df['Market_Cap'].idxmax(), 'Symbol']
```

What is the average Profit_Margin by Sector

```python
df.groupby('Sector')['Profit_Margin'].mean()
df.groupby('Sector')['Profit_Margin'].mean().plot(kind='pie', autopct='%1.1f%%')
```

Which date had the highest trading Volume

```python
df.groupby('Date')['Volume'].sum().plot(kind='line')
```

What is the maximum High price recorded

```python
df['High'].max()
df.plot(x='Date', y='High', kind='line')
```

Which stock had the lowest Low price

```python
df.groupby('Symbol')['Low'].min().plot(kind='bar')
```

What is the average Open price per Sector

```python
df.groupby('Sector')['Open'].mean().plot(kind='bar')
```

Which Symbol has the highest average Volume

```python
df.groupby('Symbol')['Volume'].mean().plot(kind='barh')
```

What is the daily price range (High − Low)

```python
df['Price_Range'] = df['High'] - df['Low']
print(df[['High', 'Low', 'Price_Range']])
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Price_Range'], marker='o')
plt.title('Daily Price Range')
plt.xlabel('Date')
plt.ylabel('Price Range (High - Low)')
plt.xticks(rotation=45)
plt.show()
```
